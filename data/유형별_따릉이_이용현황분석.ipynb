```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import csv
from matplotlib.ticker import FuncFormatter
```

```python
df = pd.read_csv('월별_따릉이_이용정보_1월.csv', encoding='cp949')
df = df[df['연령대코드'] != '기타']

# 연령대별 이용건수 총합 계산
total_by_age1 = df.groupby('연령대코드')['이용건수'].sum().reset_index()
print(total_by_age1)

# 막대그래프 시각화
plt.rc('font', family='NanumGothic')
plt.figure(figsize=(10, 6))
sns.barplot(x='연령대코드', y='이용건수', data=total_by_age1, palette='viridis')
plt.title('연령대별 이용 건수 총합(1월)')
plt.xlabel('연령대')
plt.ylabel('이용 건수 총합')
def format_fn(value, tick_number):
    return f'{int(value):,}'

plt.gca().yaxis.set_major_formatter(FuncFormatter(format_fn))
plt.show()
```

```python
df = pd.read_csv('월별_따릉이_이용정보_5월.csv', encoding='cp949')

# '\N'에 해당하는 결측치 제거
df = df[df['성별'] != '\\N']

# 대문자 F와 f를 '여성'으로, 대문자 M과 m을 '남성'으로 변환
df['성별'] = df['성별'].str.upper()
df['성별'] = df['성별'].map({'F': '여성', 'M': '남성'})

total_by_age1 = df.groupby('성별')['이동거리(M)'].mean().reset_index()
print(total_by_age1)

plt.rc('font', family='NanumGothic')
plt.figure(figsize=(10, 6))
sns.barplot(x='성별', y='이동거리(M)', data=total_by_age1, palette='viridis')
plt.title('사용자 유형별 이동거리 평균(5월)')
plt.xlabel('성별')
plt.ylabel('이동 거리 평균')
def format_fn(value, tick_number):
    return f'{int(value):,}'

plt.gca().yaxis.set_major_formatter(FuncFormatter(format_fn))
```
